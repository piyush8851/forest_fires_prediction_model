```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = pd.read_csv('/content/Algerian_forest_fires_dataset_UPDATE.csv',header=1)
```

```python
dataset.head()
```

```python
dataset.info()
```

Data Cleaning

```python
#missing value
dataset[dataset.isnull().any(axis=1)]
```

The dataset converted into 2 sets based on region from 122th index, we can make a new column based on the region :

1: **Bejaia Region Dataset**

2: **Sidi-Bel Abbes Region Dataset**

```python
dataset.loc[:122,'Region'] = 0
dataset.loc[122:,'Region'] = 1
df = dataset
```

```python
df.info()
```

```python
df.head()
```

```python
df[['Region']] = df[['Region']].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
#removing null value bcz its just 2 way less than any impact on model accuracy
df= df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
#remove the 122nd row
df = df.drop(122).reset_index(drop=True)
```

```python
df.columns
```

```python
# fix spaces in column name
df.columns = df.columns.str.strip()
df.columns
```

# Change the required column into integer data type

```python
df[['month','day','year','Temperature','RH','Ws']] = df[['month','day','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

# **Changing the other column to float datatpye**

```python
objects = [features for features in df.columns if df[features].dtypes=='O']
```

```python
for i in objects:
  if i != 'Classes':
    df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
objects
```

```python
df.describe()
```

```python
# Lets save the cleaned dataset
df.to_csv('/content/Algerian_forest_fires_cleaned_dataset.csv',index=False)
```

# **Exploratory Data Analysis**

```python
#drop date, month and year
df_copy= df.drop(['day','month','year'],axis=1)
```

```python
##plot density plot for all features
plt.style.use('seaborn-v0_8')
df_copy.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
#percentage for pie chart
percentage = df_copy['Classes'].value_counts(normalize=True)*100
```

```python
#plotting pie chart
classlabels=['Fire','Not Fire']
plt.figure(figsize=(12,7))
plt.pie(percentage,labels=classlabels,autopct='%1.1f%%')
plt.title('Pie chart of Classes')
plt.show()
```
